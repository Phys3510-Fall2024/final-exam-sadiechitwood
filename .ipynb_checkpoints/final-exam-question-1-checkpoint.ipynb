#### Pseudocode to help myself:
1. Define the potential energy function
2. Set up the equations of motion for the coupled oscillators
3. Solve the equations numerically
4. Plot the contour plot of the potential energy
5. Plot the trajectory of Q_1 and Q_2
6. Combine trajectory and contour polot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# given parameters
K1 = 1.0
K2 = 0.2
C = 0.5
M1 = 1.0
M2 = 1.0

# potential energy function
def potential_energy(Q1, Q2):
    return 0.5 * K1 * Q1**2 + 0.5 * K2 * Q2**2 + (1/3) * C * Q1 * Q2**2

# derivatives of the potential energy (forces)
def equations_of_motion(t, y):
    Q1, Q2, dQ1_dt, dQ2_dt = y
    
    # forces derived from the potential energy
    d2Q1_dt2 = -(K1 * Q1 + (1/3) * C * Q2**2) / M1
    d2Q2_dt2 = -(K2 * Q2 + (2/3) * C * Q1 * Q2) / M2
    
    return [dQ1_dt, dQ2_dt, d2Q1_dt2, d2Q2_dt2]

# initial conditions, change for physical explanation! (submit Q1_0=0, Q2_0=1)!
Q1_0 = 0
Q2_0 = 1
dQ1_dt_0 = 0
dQ2_dt_0 = 0
initial_conditions = [Q1_0, Q2_0, dQ1_dt_0, dQ2_dt_0]

# time array for the integration
t_span = (0, 50)
t_eval = np.linspace(0, 50, 200)

# solve the equations of motion, similar to previous homeworks
sol = solve_ivp(equations_of_motion, t_span, initial_conditions, t_eval=t_eval)

# extract the solutions
Q1_sol = sol.y[0]
Q2_sol = sol.y[1]

# mesh grid for the contour plot
Q1_grid, Q2_grid = np.meshgrid(np.linspace(-1.5, 1.5, 100), np.linspace(-1.5, 1.5, 100))

# element-wise operations with numpy arrays
U_grid = potential_energy(Q1_grid, Q2_grid)

# contour plot
plt.figure(figsize=(8, 6))
plt.contour(Q1_grid, Q2_grid, U_grid, levels=50, colors='white', linewidths=1)
plt.contourf(Q1_grid, Q2_grid, U_grid, levels=50, cmap='viridis')

# plot the trajectory
plt.plot(Q1_sol, Q2_sol, color='red', label="Trajectory", linewidth=1)

# labeling
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel(r'$Q_1$')
plt.ylabel(r'$Q_2$')
plt.title('Contour Plot of Potential Energy with Trajectory')
plt.legend()
plt.colorbar(label=r'$\Delta U$')
plt.grid(False) #grid made the white confusing for me
plt.show()
```

### Physical explanation
Why do the dynamics differ when only $Q_1$ is initially displaced compared to when only $Q_2$ is initially displaced? How does this relate to the coupling term in the potential energy? 1. Case 1: $Q_1(0)=0, Q_2(0)=1$
   -  This only displaced $Q_2$. Since $Q_1$ is initially at zero, the coupling term $CQ_1Q_2^2$ has no effect initially. The motion of Q_2 will be influenced only by its own spring constant of $K_2$.The trajectory may oscillate in a simple harmonic manner.


2. Case 2: $Q_1(0)=0, Q_2(0)=-1$
    - Looks similar to Case 1, but the initial displacement of $Q_2$ in the opposite direction.Behaves similarly to Case 1 because the initial conditions are symmetric in terms of displacement of $Q_2$.
      
  
3. Case 3: $Q_1(0)=1, Q_2(0)=0$
   -  Only Q_1 being displaced means the coupling term is significant. The trajectory of $Q_2$ is influencing by the coupling to $Q_1#, causing the system to show possible coupled oscillations. Remembering to plug into the `Q1_sol` and `Q2_sol` was important to see the change.
     

4. Case 4: $Q_1(0)=-1, Q_2(0)=0$
   -  This appears to just be the reverse of Case 3, as just the sign of displacement differs. The trajectory still shows coupled behavior, but starting from the negative side of $Q_1$
  


The coupling between the oscillators means that the displacemnt of one will drive the other, even if starts at rest. When initial $Q_2$ is nonzero and initial $Q_1=0$, the coupling term $\frac{1}{3}CQ_1Q_2^2$ begins to generate a force on $Q_1$, though it requires time to transfer this energy. 


The coupling term $\frac{1}{3}CQ_1Q_2^2$  is always present, but it affects the system differently depending on which is initially displaced. When $Q_1$ is diplaces, it directly affect $Q_2$. When $Q_2$ is displaced, $Q_1$ starts at rest and its motion is induced by the oscillations of $Q_2$, which is working indirectly. The asymmetic influence of the coupling terms is what makes the systems differ. 

#### Using a cylindrical beam example:
The coupling term $\frac{1}{3}CQ_1Q_2^2$  represents a local interation that couples the displactment of one part to the other. If you displace at one part, $Q_1$, creat a force on the beam that will bend the other part, $Q_2$. Over time, teh second part will start to move due to the redistribution of forces. 

If you displace $Q_2$ initally, the first part will move eventually as well, but only after $Q_2$ created enough force that the beam's properties start transferring energy. 

The effect is more immediate when $Q_1$ is displaced, and more gradual when $Q_2$ is diplaced.
