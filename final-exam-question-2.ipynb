### FUNCTION 1: POTENTIAL ENERGY

```python
def potential_energy(Q, K=1, D=2.4, Phi=1):
    return (1/2) * K * Q**2 - (1/4) * D * Q**4 + (1/6) * Phi * Q**6

# test for potential_energy
assert np.isclose(potential_energy(0), 0)  # expected output for Q=0
```

### FUNCTION 2: SECOND DERIVATIVE OF POTENTIAL ENERGY

```python
def d2U_dQ2(Q, K=1, D=2.4, Phi=1):
    return K - D * Q**2 + Phi * Q**4

# tests for d2U_dQ2
assert np.isclose(d2U_dQ2(0), 1)  # second derivative at Q=0 should be K
assert np.isclose(d2U_dQ2(1), -0.4)  # check at another point
```

### FUNCTION 3: NUMERICAL SOLVER (EULER'S METHOD)

```python
def solve_differential_eq(Q0, M=1, T_max=20, time_steps=1000):
    dt = T_max / time_steps
    Q = np.zeros(time_steps)
    Q[0] = Q0
    velocity = 0  # Uinitial velocity is zero

    for t in range(1, time_steps):
        force = -d2U_dQ2(Q[t-1])  # F = -dU/dQ
        acceleration = force / M
        velocity += acceleration * dt
        Q[t] = Q[t-1] + velocity * dt
    return np.linspace(0, T_max, time_steps), Q

# test for solve_differential_eq (using a small time step and known initial condition)
t, Q = solve_differential_eq(1)
assert len(t) == 1000  # test if the time array has the expected length
```

### FUNCTION 4: IDENTIFY THE PEAKS

```python
def find_peaks(Q):
    peaks = []
    for i in range(1, len(Q)-1):
        if Q[i-1] < Q[i] and Q[i] > Q[i+1]:
            peaks.append(i)
    return peaks

# test for find_peaks
assert len(find_peaks([0, 1, 0])) == 1  # peak should be at index 1
```

### FUNCTION 5: PERIOD CALCULATION

```python
def calculate_period(Q0):
    t, Q = solve_differential_eq(Q0)
    peaks = find_peaks(Q)
    if len(peaks) >= 2:
        return t[peaks[1]] - t[peaks[0]]  # return the time between the first two peaks
    return None

# test for calculate_period
T = calculate_period(1)
assert T is not None  # period is found!
assert T > 0  # period should be positive
```

### FUNCTION 6: ANGULAR FREQUENCY CALCULATION

```python
def calculate_angular_frequency(T):
    return 2 * np.pi / T

# test for calculate_angular_frequency
assert np.isclose(calculate_angular_frequency(2), np.pi)  # expected output for T=2
```
